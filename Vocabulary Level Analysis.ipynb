### Komoran 기준
- 감 : 감탄사 IC 
- 고 : 고유 명사 NNP	
- 관 : 관형사 MM
- 대 : 대명사 NP
- 동 : 동사 VV
- 명 : 명사 NNG
- 보 : 보조 용언	VX
- 부 : 부사 MAG
- 불 : 분석 불능 -> 제거하기
- 수 : 수사 NR
