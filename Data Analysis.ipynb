```python
from konlpy.tag import Kkma 
from konlpy.utils import pprint
kkma = Kkma()
```

```python
# 문장 별 평균 길이 반환 함수 
def sent_len(desc):
    sent_list = kkma.sentences(desc)
    sum_length =0
    for i in range(len(sent_list)):
        sum_length+=len(sent_list[i])
    return(sum_length/len(sent_list))
```

```python
sent_len(full.loc[0]['desc'])
```

```python
full['sent_len'] = full['desc'].apply(lambda x: sent_len(x))
```

```python
full.tail()
```

```python
# 명사 개수 / 전체 글 길이
def noun(desc):
    return len(kkma.nouns(desc)) / len(desc)
```

```python
full['noun'] = full['desc'].apply(lambda x: noun(x))
```

```python
# 종결 어미의 개수 / 문장의 길이
# 'EFN', 'EFQ', 'EFO','EFA','EFI','EFR'
# 각각 평서, 의문, 명령, 청유, 감탄, 존칭 

def ending(desc):
    n = list(dict(kkma.pos(desc)).values()).count('EFN')
    q = list(dict(kkma.pos(desc)).values()).count('EFQ')
    o = list(dict(kkma.pos(desc)).values()).count('EFO')
    a = list(dict(kkma.pos(desc)).values()).count('EFA')
    i = list(dict(kkma.pos(desc)).values()).count('EFI')
    r = list(dict(kkma.pos(desc)).values()).count('EFR')
    return((n+q+o+a+i+r)/ len(desc))
```

```python
import pandas as pd 
df = pd.read_csv('movie_project_factors.csv')
```

```python
df['ending'] = df['desc'].apply(lambda x: ending(x))
```

```python
# 연결 어미 개수 / 전체 글 길이
# 'ECE' 대등
# 'ECS' 보조
# 'ECD' 의존 
def connect(desc):
    e = list(dict(kkma.pos(desc)).values()).count('ECE')
    s = list(dict(kkma.pos(desc)).values()).count('ECS')
    d = list(dict(kkma.pos(desc)).values()).count('ECD')
    return((e+s+d)/ len(desc))
```

```python
df.groupby('media')['connect'].describe()
```

```python
full['connect'] = full['desc'].apply(lambda x: connect(x))
```

```python
# 관형형 전성어미 개수 / 전체 글 길이 
def determ(desc):
    etd = list(dict(kkma.pos(desc)).values()).count('ETD')
    return(etd/ len(desc))
```

```python
full['determ'] = full['desc'].apply(lambda x: determ(x))
```

```python
# 형용사 개수 / 전체 글 길이 
def adj(desc):
    adj = list(dict(kkma.pos(desc)).values()).count('VA')
    return(adj/len(desc))
```

```python
full['adj'] = full['desc'].apply(lambda x: adj(x))
```

```python
# 동사 개수 / 전체 글 길이 
def verb(desc):
    verb = list(dict(kkma.pos(desc)).values()).count('VV')
    return(verb/len(desc))
```

```python
full['verb'] = full['desc'].apply(lambda x: verb(x))
```

```python
# 대명사 개수 / 전체 글 길이 
def pronoun(desc):
    pronoun = list(dict(kkma.pos(desc)).values()).count('NP')
    return(pronoun/len(desc))
```

```python
full['pronoun'] = full['desc'].apply(lambda x: pronoun(x))
```

```python
# 부사 개수 / 전체 글 길이 
def adverb(desc):
    adverb = list(dict(kkma.pos(desc)).values()).count('MAG')
    return(adverb/len(desc))
```

```python
df['adverb'] = df['desc'].apply(lambda x: adverb(x))
```

```python
# 접속 부사 개수 / 전체 글 길이
def conj(desc):
    conj = list(dict(kkma.pos(desc)).values()).count('MAC')
    return(conj/len(desc))
```

```python
df['conj'] = df['desc'].apply(lambda x: conj(x))
```
