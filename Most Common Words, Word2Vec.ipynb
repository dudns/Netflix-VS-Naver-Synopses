# Most Common Words

```python
import pandas as pd 
df = pd.read_csv('pos.csv')
```

```python
with open('netflix.txt', 'r', encoding='UTF8') as f:
    lines = f.read().splitlines()
lines
```

```python
sentences = [line for line in lines if line != '']
```

```python
from konlpy.tag import Komoran
tagger = Komoran()
tagged_sentences = [tagger.pos(sent) for sent in sentences]
```

```python
word_list = []
for sent in tagged_sentences:    
    for word in sent:
            word_list.append(word)
```

```python
from collections import Counter
word_counts = Counter(word_list)
word_counts.most_common()
```

```python
netflix_freq = pd.DataFrame(word_counts.most_common(), columns=['word', 'freq'])
netflix_freq
```

```python
netflix_freq.to_csv('netflix_freq.csv', index=False, encoding='utf-8')
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
split = [kkma.morphs(sent) for sent in sentences]
```

```python
model = Word2Vec(split, size=100, window=5, min_count=20, workers=4, iter=50, sg=1)
```

```python
word_vectors = model.wv
vocabs = word_vectors.vocab.keys()
word_vectors_list = [word_vectors[v] for v in vocabs]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
xys = pca.fit_transform(word_vectors_list)
xs = xys[:,0]
ys=xys[:,1]
```

```python
import matplotlib.pyplot as plt

def plot_2d_graph(vocabs, xs, ys):
    plt.figure(figsize=(15,10))
    plt.scatter(xs,ys,marker='o')
    for i,v in enumerate(vocabs):
        plt.annotate(v,xy=(xs[i], ys[i]))
```
