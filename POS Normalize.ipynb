```python
from konlpy.tag import Kkma
kkma = Kkma()

# 문장 개수 반환 함수
def sent_num(desc):
    sent_list = kkma.sentences(desc)
    return(len(sent_list))
```

```python
df['sent_num'] = df['desc'].apply(lambda x: sent_num(x))
```

```python
df.head()
```

```python
df.iloc[:,6:-1]
```

```python
df.iloc[:, 6:-1]  = df.iloc[:, 6:-1] .apply(lambda x: x/df['sent_num'])
```

```python
df.head()
```

```python
df.to_csv('pos_normalized.csv', index=False, encoding='utf-8')
```
